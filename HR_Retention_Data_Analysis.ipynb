## Model

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
cols_to_transform = ['salary','department']
df = pd.get_dummies(df,columns = cols_to_transform )
```

```python
df.head()
```

```python
x = df.drop(['left'], axis=1)
y = df['left']
```

```python
x.head()
```

```python
#Scaling the data in preparation for the model 
#scaler = StandardScaler()
#x = scaler.fit_transform(x)
x = pd.DataFrame(StandardScaler().fit_transform(x), columns=x.columns, index=x.index)
```

```python
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0, stratify=y)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_prob = model.predict_proba(x)[:,1]
```

```python
df_pred = df
df_pred['Leave Prediction'] = model.predict(x)
df_pred['Leave Probability'] = y_prob

df_pred.head()
```

```python
print("Accuracy: ", model.score(x_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
#Evaluating model performance 
confusion_matrix(y_test,y_pred)
```

```python
# To find the weights of each dimension
weights = pd.Series(model.coef_[0], index=x.columns.values)
weights.sort_values(ascending=False)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Categorise employees into low, 
df_pred['Risk'] = pd.cut(x=df_pred['Leave Probability'], bins=[0,0.4,0.75,1.0], labels=['Low', 'Medium','High'])
df_pred.head()
```
